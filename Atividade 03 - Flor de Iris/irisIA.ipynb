```python
# Checando os dados da distribui√ß√£o
plt.figure(figsize=(10,8))
sns.distplot(iris['sepal_length'], fit=stats.norm, kde = False)
plt.show()
```

```python
# Teste de Normalidade de Shapiro-Wilk
shapiro_stat, shapiro_p = stats.shapiro(iris['sepal_length'])
shapiro_p

if shapiro_p > 0.05:
  print("A distribui√ß√£o √© normal")
else:
  print("A distribui√ß√£o n√£o √© normal")
```

```python
# Gr√°fico de Dispers√£o
plt.figure(figsize=(12,6))
sns.regplot(x='sepal_length', y='sepal_width', data=iris, ci=None) # Regress√£o Linear

# Adicionando os Labels
plt.title("Gr√°fico de Dispers√£o - Flor de √çris")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```
