```python
# Análise de Dados com o Histograma
iris.hist(grid=False, figsize=(10,10))
plt.show()
```

```python
# Checando os dados da distribuição
plt.figure(figsize=(10,8))
sns.distplot(iris['sepal_length'], fit=stats.norm, kde = False)
plt.show()
```

```python
# Teste de Normalidade de Shapiro-Wilk
shapiro_stat, shapiro_p = stats.shapiro(iris['sepal_length'])
shapiro_p

if shapiro_p > 0.05:
  print("A distribuição é normal")
else:
  print("A distribuição não é normal")
```

```python
# Gráfico de Dispersão
plt.figure(figsize=(12,6))
sns.regplot(x='sepal_length', y='sepal_width', data=iris, ci=None) # Regressão Linear

# Adicionando os Labels
plt.title("Gráfico de Dispersão - Flor de Íris")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
# Regressão Linear
stats.linregress(iris['sepal_length'], iris['sepal_width'])
```
